```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
training_set_feat = pd.read_csv('training_set_features.csv', index_col = "respondent_id")
training_set_label = pd.read_csv('training_set_labels.csv',index_col = "respondent_id")
test_set_feat = pd.read_csv('test_set_features.csv',index_col = "respondent_id")
submission_format = pd.read_csv('submission_format.csv')
```

```python
training_joined = training_set_feat.join(training_set_label)
```

```python
sns.countplot(x="h1n1_vaccine", data = training_joined )
```

```python
training_joined.drop(columns=["hhs_geo_region","employment_industry","employment_occupation","seasonal_vaccine"], inplace = True)
```

```python
for col in training_joined[training_joined.dtypes[training_joined.dtypes == "object"].index].columns:
    sns.set(rc={'figure.figsize':(9,5)})
    sns.countplot(y = training_joined[col], hue=training_joined["h1n1_vaccine"])
    plt.title( col +" Countplot")
    plt.xticks(rotation=90)
    plt.show()
```

```python
for col in training_joined[training_joined.dtypes[training_joined.dtypes != "object"].index].columns:
    sns.set(rc={'figure.figsize':(9,5)})
    sns.countplot(y = training_joined[col], hue=training_joined["h1n1_vaccine"])
    plt.title( col +" Countplot")
    plt.xticks(rotation=90)
    plt.show()
```
